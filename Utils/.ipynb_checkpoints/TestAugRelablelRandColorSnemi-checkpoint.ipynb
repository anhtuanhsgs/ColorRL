```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
import albumentations as A
import cv2
import colorsys
```

```python
raw_list = io.imread ("../Data/snemi/train/A/ac4_images_cc.tif")
lbl_list = io.imread ("../Data/snemi/train/B/train-labels.tif")
```

```python
def aug (image, mask):
    aug = A.Compose([
                A.HorizontalFlip (p=0.5),
                A.VerticalFlip(p=0.5),              
                A.RandomRotate90(p=0.5),
                A.Transpose (p=0.5),
                A.OneOf([
                    A.ElasticTransform(p=0.5, alpha=1, sigma=5, alpha_affine=5, interpolation=cv2.INTER_NEAREST, border_mode=cv2.BORDER_CONSTANT),
                    A.GridDistortion(p=0.5, interpolation=cv2.INTER_NEAREST, border_mode=cv2.BORDER_CONSTANT),
                    A.OpticalDistortion(p=0.5, distort_limit=(0.2, 0.2), shift_limit=(0, 0), interpolation=cv2.INTER_NEAREST, border_mode=cv2.BORDER_CONSTANT),                 
                    ], p=0.6),
                A.ShiftScaleRotate (p=0.5, shift_limit=0.15, rotate_limit=45, interpolation=cv2.INTER_NEAREST, scale_limit=(0.1, 0.1), border_mode=cv2.BORDER_CONSTANT),
                # A.CLAHE(p=0.3),
                A.RandomBrightness (p=0.9, limit=0.5),
                A.RandomContrast (p=0.5),
                A.GaussNoise (p=0.5),
                A.Blur (p=0.3, blur_limit=4),
#                 A.RandomCrop (256, 256),
                ]
            )

    ret = aug (image=image, mask=mask)
    return ret ['image'], ret ['mask']
```

```python
img_id = 44
raw, lbl = raw_list [img_id], lbl_list [img_id]
```

```python
raw, lbl = raw_list [img_id], lbl_list [img_id]
fig=plt.figure(figsize=(8, 8))
fig.add_subplot (2, 2, 1)
plt.imshow (raw, cmap='gray')
fig.add_subplot (2, 2, 2)
plt.imshow (lbl, cmap='tab20')
raw, lbl = aug (raw, lbl)
fig.add_subplot (2, 2, 3)
plt.imshow (raw, cmap='gray')
fig.add_subplot (2, 2, 4)
plt.imshow (lbl, cmap='tab20')
plt.show ()
```

```python
def color_generator (N):
#   (x%3)*0.2 + 0.5
    HSV_tuples = [(x*1.0/N, 0.5, ) for x in range(N)]
    RGB_tuples = list (map(lambda x: colorsys.hsv_to_rgb(*x), HSV_tuples))

    COLOR_LIST = [(int (rgb[0] * 255), int (rgb[1] * 255), int (rgb[2] * 255)) for rgb in RGB_tuples]
    # shuffle (COLOR_LIST)
    COLOR_LIST [0] = (0, 0, 0)

    def index2rgb (index):
        return COLOR_LIST [index]
    def lbl2rgb (lbl):
        original_shape = np.squeeze (lbl).shape
        lbl = lbl.flatten ().tolist ()
        lbl = list (map (index2rgb, lbl))
        lbl = np.array (lbl).reshape (original_shape + (3,))
        return lbl
    return lbl2rgb
```

```python
N = 111
lbl2rgb = color_generator (111)
```

```python
color_list = list (range (120))
color_list = np.array (color_list).reshape(10, 12)
color_list = color_list * (color_list < 111)

print (np.max (color_list), np.min (color_list))

color_list = lbl2rgb (color_list)
plt.imshow (color_list)
plt.show ()
```
